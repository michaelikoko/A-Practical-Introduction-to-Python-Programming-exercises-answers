```python
#1
'''
Write a class called Investment with fields called principal and interest. The constructor should set the values of those fields. There should be a method called 
value_after that returns the value of the investment after n years. The formula for this is p(1 + i)^n, where p is the principal, and i is the interest rate. It 
should also use the special method __str__ so that printing the object will result in something like below:
    Principal - $1000.00, Interest rate - 5.12%
'''
class Investment:
    def __init__(self, p, i):
        self.p = p
        self.i = i

    def __str__(self):
        return f'Principal - ${self.p:.2f}, Interest rate - {self.i:>3.2f}%'

    def valueAfter(self, n):
        return self.p*(1 + self.i)**n

investment = Investment(float(input('Input principal($): ')), float(input('Input the interest rate(%): ')))
print(investment)
n = eval(input('Input the number of years(n): '))
print(f'The value of the investment after {n} years, is ${investment.valueAfter(n):.2f}')
```

```python
#2
'''
Write a class called Product. The class should have fields called name, amount, and price, holding the product’s name, the number of items of that product in stock, 
and the regular price of the product. There should be a method get_price that receives the number of items to be bought and returns a the cost of buying that many 
items, where the regular price is charged for orders of less than 10 items, a 10% discount is applied for orders of between 10 and 99 items, and a 20% discount is 
applied for orders of 100 or more items. There should also be a method called make_purchase that receives the number of items to be bought and
decreases amount by that much.
'''
class Product:
    def __init__(self, name, amount, price):
        self.name = name
        self.amount = amount
        self.price = price

    def getPrice(self, n):
        if n < 10: return n * self.price
        elif 10 <= n <= 99: return n * (10/100 * self.price)
        elif n > 99: return n * (20/100 * self.price)

    def makePurchase(self, n):
        self.amount -= n
name = 'Coconut'; amount = 500; price = 250
product = Product(name, amount, price)
n = int(input('Input the number of items to be bought: '))
print(f'The cost of buying {n} {name.title()}s is N{product.getPrice(n):.2f}')
product.makePurchase(n)
```

```python
#4
'''
Write a class called Time whose only field is a time in seconds. It should have a method called convert_to_minutes that returns a string of minutes and seconds 
formatted as in the following example: if seconds is 230, the method should return '5:50'. It should also have a method called convert_to_hours that returns a 
string of hours, minutes, and seconds formatted analogously to the previous method.
'''
class Time:
    def __init__(self, sec):
        self.sec = sec

    def convertToMinutes(self):
        min = self.sec // 60; remSec = self.sec % 60
        if min < 10: min = f'0{min}'
        if remSec < 10: remSec = f'0{remSec}'
        return f'{min}:{remSec}'

    def convertToHours(self):
        min = self.sec // 60
        hour = min // 60; m = min % 60; remSec = self.sec % 60
        if m < 10: m = f'0{m}'
        if remSec < 10: remSec = f'0{remSec}'
        if hour < 10: hour = f'0{hour}'
        return f'{hour}:{m}:{remSec}'

time = Time(int(input('Input time in seconds: ')))
print(time.convertToMinutes()); print(time.convertToHours())
```

```python
#5
'''
Write a class called Wordplay. It should have a field that holds a list of words. The user of the class should pass the list of words they want to use to the class. 
There should be the following methods:
    • words_with_length(length) — returns a list of all the words of length length
    • starts_with(s) — returns a list of all the words that start with s
    • ends_with(s) — returns a list of all the words that end with s
    • palindromes() — returns a list of all the palindromes in the list
    • only(L) — returns a list of the words that contain only those letters in L
    • avoids(L) — returns a list of the words that contain none of the letters in L
'''
class Wordplay:
    def __init__(self, wordList):
        self.wordList = []
        for word in wordList:
            word = word.strip()
            word = word.lower()
            self.wordList.append(word)

    def wordsWithLength(self, length):
        l = [word for word in self.wordList if len(word) == length]
        return l

    def startsWith(self, s):
        l = [word for word in self.wordList if word.startswith(s.lower())]
        return l

    def endsWith(self, s):
        l = [word for word in self.wordList if word[-(len(s)):] == s.lower()]
        return l

    def palindromes(self):
        l = [word for word in self.wordList if word == word[::-1]]
        return l

    def only(self, letterList):
        l = [word for word in self.wordList if sorted(list(word)) == sorted(letterList)]
        return l

    def avoids(self, letterList):
        l = []
        for word in self.wordList:
            for letter in list(word):
                if letter in letterList:
                    break
            else:
                l.append(word)
        return l
wordList = input('Input a list of words separated by a comma (","): ').split(',')
wordPlay = Wordplay(wordList)
n = int(input('Input word length: '))
print(f'Words with length {n}: {wordPlay.wordsWithLength(n)}')
s = input('Input a letter: ')
print(f'Words that starts with the letter {s}: {wordPlay.startsWith(s)}')
print(f'Words that ends with the letter {s}: {wordPlay.endsWith(s)}')
print(f'Words that are palindromes: {wordPlay.palindromes()}')
L = input('Input a list of letters separated by a comma (","): ').split(',')
letterList = []
for l in L:
    l = l.lower()
    l = l.strip()
    letterList.append(l)
print(f'Words that only contain the letters {letterList}: {wordPlay.only(letterList)}')
print(f'Words that contain none of the letters {letterList}: {wordPlay.avoids(letterList)}')
```

```python
#7
'''
Use the Standard_deck class of this section to create a simplified version of the game War. In this game, there are two players. Each starts with half of a deck. 
The players each deal the top card from their decks and whoever has the higher card wins the other player’s cards and adds them to the bottom of his deck. If there
is a tie, the two cards are eliminated from play (this differs from the actual game, but is simpler to program). The game ends when one player runs out of cards.
'''
import random
class Card:
    def __init__(self, value, suit):
        self.value = value
        self.suit = suit

    def __str__(self):
        names = ['Jack', 'Queen', 'King', 'Ace']
        if self.value <= 10:
            return f'{self.value} of {self.suit}'
        else:
            return f'{names[self.value-11]} of {self.suit}'

class Card_game:
    def __init__(self, cards = []):
        self.cards = cards

    def shuffle_cards(self):
        random.shuffle(self.cards)

    def players(self):
        p1 = self.cards[:len(self.cards)//2]
        p2 = self.cards[len(self.cards)//2:]
        return p1, p2

    def deal(self, p1, p2):
        return p1.pop(0), p2.pop(0)

    def check(self, c1, c2):
        if c1 > c2:
            return 1
        elif c1 < c2:
            return 2
        elif c1 == c2:
            return 0

class Standard_deck(Card_game):
    def __init__(self):
        self.cards = []
        for s in ['Hearts', 'Diamonds', 'Clubs', 'Spades']:
            for v in range(2, 15):
                self.cards.append(Card(v, s))

game = Standard_deck()
game.shuffle_cards()
player1, player2 = game.players()
while True:
    card1, card2 = game.deal(player1, player2); print(f'Player 1: {card1}, Player 2: {card2}')
    cond = game.check(card1.value, card2.value)
    if cond == 1:
        player1.append(card2)
        print('Player 1 wins this round.')
    elif cond == 2:
        player2.append(card1)
        print('Player 2 wins this round.')
    elif cond == 0:
        print("It is a tie.")
    if len(player1) == 0:
        print('Player 2 wins.')
        break
    elif len(player2) == 0:
        print('Player 1 wins.')
        break
    
```

```python
#We start with a class for a playing card. The data associated with a card consists of its value (2 through
# 14) and its suit. The Card class below has only one method __str__. This is a special method that, among
#other things, tells the print statement how to print a Card object.

class Card:
    def __init__(self, value, suit):
        self.value = value
        self.suit = suit

    def __str__(self):
        names = ['Jack', 'Queen', 'King', 'Ace']
        if self.value <= 10:
            return f'{self.value} of {self.suit}'
        else:
            return f'{names[self.value-11]} of {self.suit}'

#Next we have a class to represent a group of cards. Its data consists of a list of Card objects. It has a 
#number of methods: nextCard which removes the first card from the list and returns it; hasCard which returns
#True or False depending on if there are any cards left in the list; size which returns how many cards are in
#the list; and shuffle, which shuffles the list.
import random

class Card_group:
    def __init__(self, cards=[]):
        self.cards = cards

    def nextCard(self):
        return self.cards.pop(0)

    def hasCard(self):
        return len(self.cards) > 0

    def size(self):
        return len(self.cards)

    def shuffle(self):
        random.shuffle(self.cards)

#We have one more class Standard_deck, which inherits from Card_group. The idea here is that Card_group 
#represents an arbitary group of cards, and Standard_deck represents a specific group of cards, namely the 
#standard deck of 52 cards used in most card games.

class Standard_deck(Card_group):
    def __init__(self):
        self.cards = []
        for s in ['Hearts', 'Diamonds', 'Clubs', 'Spades']:
            for v in range(2, 15):
                self.cards.append(Card(v, s))

#One way to think of what we have done with the classes is that we have built up a miniature card programming
#language, where we can think about how a card game works and not have to worry about exactly how cards are
#shuffled or dealt or whatever, since that is wrapped up into the classes. For the hi-low game, we get a new
#deck of cards, shuffle it. A nice feature of this game is that it deals out all 52 cards of a deck, so a player
#can use their memory to help them play the game.
deck = Standard_deck()
deck.shuffle()

new_card = deck.nextCard()
print(f'\n {new_card}')
choice = input('Higher (h) or lower (l): ')
streak = 0

while (choice == 'h' or choice == 'l'):
    if not deck.hasCard():
        deck = Standard_deck()
        deck.shuffle()

    old_card = new_card
    new_card = deck.nextCard()

    if (choice.lower() == 'h' and new_card.value > old_card.value or choice.lower() == 'l' and new_card.value <\
        old_card.value):
        streak += 1
        print(f'Right! That\'s {streak} in a row.')
    elif (choice.lower() == 'h' and new_card.value < old_card.value or choice.lower() == 'l' and new_card.value\
        > old_card.value):
        streak = 0
        print('Wrong.')
    else:
        print('Push.')
    print(f'\n {new_card}')
    choice = input('Higher (h) or lower (l): ')
```

```python
#9
'''
Write a class called Rock_paper_scissors that implements the logic of the game Rockpaper-scissors. For this game the user plays against the computer for a certain
number of rounds. Your class should have fields for the how many rounds there will be, the current round number, and the number of wins each player has. There 
should be methods for getting the computer’s choice, finding the winner of a round, and checking to see if someone has won
the (entire) game. You may want more methods.
'''
import random
class Rps:
    def __init__(self, numRounds):
        self.numRounds = numRounds
        self.curRound = 0
        self.winUser = 0
        self.winComp = 0

    def computerChoice(self):
        return random.randint(1, 3)

    def checkWinner(self, user, comp, name):
        check = [[0, -1, 1], [1, 0, -1], [-1, 1, 0]]
        self.curRound += 1
        if check[user-1][comp-1] == 0:
            self.winUser += 1
            self.winComp += 1
            return f'It is a tie.\nSCORES {name}: {self.winUser} | Computer: {self.winComp}'
        elif check[user-1][comp-1] == 1:
            self.winUser += 1
            return f'{name} wins.\nSCORES {name}: {self.winUser} | Computer: {self.winComp}'
        elif check[user-1][comp-1] == -1:
            self.winComp += 1
            return f'Computer wins.\nSCORES {name}: {self.winUser} | Computer: {self.winComp}'

    def endGame(self):
        if self.curRound == self.numRounds: 
            return False
        return True

    def wonGame(self, name):
        if self.winComp > self.winUser:
            return f'Game over. {name} loses.'
        elif self.winComp < self.winUser:
            return f'Good job! {name} win.'
        elif self.winUser == self.winComp:
            return 'The game ends in a draw.'
l = ['Rock', 'Paper', 'Scissors']
name = input('Input name: ')
if name == '':
    name = 'User'
name = name.title()
while True:
    while True:
        try:
            n = int(input('Input number of rounds: '))
            break
        except:
            print('Please input an integer.')
    game = Rps(n)
    while game.endGame():
        while True:
            try:
                print(f'\nRound {game.curRound+1}:')
                user_action = int(input('\nRock(1), Paper(2), Scissors(3): '))
                if user_action in [1, 2, 3]:
                    break
                else:
                    print('You can only input numbers between 1 and 3.')
            except:
                print('Input an integer ')
        comp = game.computerChoice()
        print(f'{name}: {l[user_action-1]} | Computer: {l[comp-1]}')
        print(game.checkWinner(user_action, comp, name))
    print(game.wonGame(name))
    print('Do you want to play again? (yes or no)')
    if not input('> ').lower().startswith('y'):
        break
print('Thanks for playing!')
```
